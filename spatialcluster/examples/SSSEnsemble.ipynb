```python
import pygeostat as gs
% matplotlib inline
```

```python
import spatialcluster as sp
```

```python
custom = {"savefig.dpi": 125,
            "figure.dpi": 125,
            "figure.figsize": (4, 3),
            "axes.grid": True,
            'grid.alpha': 1.0,
            'grid.color': 'black',
            'grid.linestyle': '-.',
            'grid.linewidth': 0.25,
            "axes.formatter.useoffset": False,
            "xtick.major.pad": 1,
            "ytick.major.pad": 1}
gs.set_style("pt7", custom)
```

```python
datfl = gs.DataFile("testdat.dat")
```

```python
gs.locmap(datfl, var='v1', figsize=(6, 6))
```

# first generate some codes to train against

```python
mvdata = datfl.data[['v1', 'v2', 'v3', 'v4']].values
```

```python
locations = datfl[['X', 'Y']].values
```

```python
dss = sp.DSSEnsemble(mvdata, locations, nreal=100, nnears=50, numtake=1)
```

```python
dss.fit(6)
```

```python
datfl.data['final'] = dss.predict(4, method='spec')
gs.locmap(datfl, var='final', cmap='jet', figsize=(6, 6))
```

# now generate classess with the ssdss method

```python
ssdss = sp.SSDSSEnsemble(mvdata, locations, nreal=100, nnears=10, numtake=10, 
                         domaincodes=datfl['final'], seedprop=0.25)
```

```python
ssdss.fit()
```

```python
datfl.data['final_ss'] = ssdss.predict(method='spec')
```

```python
gs.locmap(datfl, var='final_ss', title='SemiSupervised', cmap='jet', figsize=(6, 6))
```

```python
gs.locmap(datfl, var='final', title='Spatial Clusters', cmap='jet', figsize=(6, 6))
```
